```python
proj('hgrid.gr3',0,'epsg:4326','hgrid_cart.gr3',0,'epsg:3857')

gd=read('hgrid_cart.gr3');
gd.compute_all()

import numpy as np
import pandas as pd

time_stamps_sim_0 = pd.date_range(start='2018-11-1 00:00:00', end='2019-5-31 23:00:00', freq="6h")
time_stamps_sim = pd.date_range(start='2019-6-1 00:00:00', end='2019-6-22 23:00:00', freq="6h")
num_file = len(time_stamps_sim)
file_start = len(time_stamps_sim_0)
# print(num_file)
# time_arr = np.zeros(num_file)
for time in range(file_start, min(file_start+num_file, 954), 1):
    file_ind = time+1
    # time_arr[time] = time*6.0*3600
    filename = "./outputs/out2d_"+str(file_ind)+".nc"
    ncfile=ReadNC(filename)
    if time==0:
        ave_thick_arr = ncfile.iceTracer_1.val
        ave_frac_arr = ncfile.iceTracer_2.val
    else:
        ave_thick_arr += ncfile.iceTracer_1.val
        ave_frac_arr += ncfile.iceTracer_2.val
    print(time)

ave_thick_arr/=len(time_stamps_sim)
ave_frac_arr/=len(time_stamps_sim)

C=ReadNC(r'./outputs/monthly-ave/nov.nc')         #read data
C.iceTracer_1.val=ave_thick_arr                     #modify value
C.iceTracer_2.val=ave_frac_arr                     #modify value
WriteNC(r'./outputs/monthly-ave/out2d_8.nc',C) #write a new file
```
